The Number of Male and Female employes in the company 

```python
ax = sns.countplot(data = df, x= "Sex")
for bars in ax.containers:
    ax.bar_label(bars)
```

Marital statius of the employees in the company 

```python
ax = sns.countplot(data = df, x= "MaritalDesc",hue = "Sex")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Average salary of the employees on the basis od departments  
```

```python
dep= df.groupby(['Department'],as_index = False,)['Salary'].mean().sort_values(by ="Salary",ascending = False)


sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data= dep,x='Department',y ='Salary')
```

Avegave saticfaction of the employees in different departments

```python
de= df.groupby(['Department'],as_index = False,)['EmpSatisfaction'].mean()



sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data= de,x='Department',y ='EmpSatisfaction')
```
